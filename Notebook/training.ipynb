```python
# Some imports
import os
import cv2
import numpy as np

from PIL import Image
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
root = ".."
img_dir = "images"
img_name = "moon_1.jpg"
img_path = os.path.join(root, img_dir, img_name)
```

```python
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img, cmap='gray')
```

```python
inverted = cv2.bitwise_not(img)
plt.imshow(inverted, cmap='gray')
```

```python
mean_org = img.mean()
mean_inverted = inverted.mean()
```

```python
img[img > 50] = 255
plt.imshow(img, cmap='gray')
```

```python
inverted[inverted > 100] = 255
plt.imshow(inverted, cmap='gray')
```

```python
new_img = (img) + (inverted)
plt.imshow(new_img, cmap='gray')
```

```python
new_img = new_img.astype(np.uint8)
circles = cv2.HoughCircles(new_img, cv2.HOUGH_GRADIENT, 1, 1, param1=50, param2=30, minRadius=0, maxRadius=0)
```

```python
circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(new_img, (i[0], i[1]), i[2], (0, 255, 0), 2)
    # draw the center of the circle
    cv2.circle(new_img, (i[0], i[1]), 2, (255, 0, 0), 3)

plt.imshow(new_img, cmap='gray')
```
